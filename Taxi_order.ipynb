#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

## Описание проекта

### Цель проекта


Разработать модель машинного обучения для прогнозирования количества заказов такси в аэропортах на следующий час с точностью, соответствующей значению RMSE ≤ 48 на тестовой выборке. Это позволит сервису «Чётенькое такси» эффективно управлять пиковой нагрузкой, привлекая водителей заранее.

## Анализ ( Domain-specific )

Меня посетила идея создать отдельный датафрейм, чтобы дополнить признаками исследование для анализа.

### Создание датафрейма для анализа

```python
df_new = data_taxi_hourly_sum.copy()

df_new['hour'] = df_new.index.hour
df_new['month'] = df_new.index.month
df_new['day_of_week'] = df_new.index.dayofweek
df_new['date_only'] = df_new.index.date
df_new['day_of_month'] = df_new.index.day

day_names_dict = {0: 'Понедельник', 1: 'Вторник', 2: 'Среда',
                  3: 'Четверг', 4: 'Пятница', 5: 'Суббота', 6: 'Воскресенье'}
day_names_short = {0: 'Пн', 1: 'Вт', 2: 'Ср', 3: 'Чт', 4: 'Пт', 5: 'Сб', 6: 'Вс'}

df_new['day_name'] = df_new['day_of_week'].map(day_names_dict)
df_new['day_name_short'] = df_new['day_of_week'].map(day_names_short)

df_new['is_weekend'] = df_new['day_of_week'].isin([5, 6]).astype(int)
df_new['day_type'] = np.where(df_new['is_weekend'] == 1, 'Выходной', 'Будний')

holidays_2018 = {
    '2018-03-08': 'Международный женский день',
    '2018-05-01': 'Праздник Весны и Труда',
    '2018-05-02': 'Майские праздники',
    '2018-05-09': 'День Победы',
    '2018-06-11': 'День России',
    '2018-06-12': 'День России (выходной)'
}

df_new['is_holiday'] = 0
df_new['holiday_name'] = ''

for holiday_date, holiday_name in holidays_2018.items():
    holiday_dt = datetime.strptime(holiday_date, '%Y-%m-%d').date()
    mask = df_new['date_only'] == holiday_dt
    df_new.loc[mask, 'is_holiday'] = 1
    df_new.loc[mask, 'holiday_name'] = holiday_name

df_new.loc[df_new['is_holiday'] == 1, 'day_type'] = 'Праздник'

def get_day_category(row):
    if row['is_holiday'] == 1:
        return 'Праздник'
    elif row['is_weekend'] == 1:
        return 'Выходной'
    else:
        return 'Будний'

df_new['day_category'] = df_new.apply(get_day_category, axis=1)

daily_aggregated = df_new.groupby('date_only').agg({
    'num_orders': 'sum',
    'day_name': 'first',
    'day_name_short': 'first',
    'day_of_week': 'first',
    'is_weekend': 'first',
    'is_holiday': 'first',
    'holiday_name': 'first',
    'day_type': 'first',
    'day_category': 'first'
}).reset_index()

daily_aggregated.columns = [
    'Дата',
    'Количество_заказов',
    'День_недели',
    'День_недели_коротко',
    'День_недели_число',
    'Выходной',
    'Праздник',
    'Название_праздника',
    'Тип_дня',
    'Категория_дня'
]

daily_aggregated['Месяц'] = pd.to_datetime(daily_aggregated['Дата']).dt.month
daily_aggregated['День_месяца'] = pd.to_datetime(daily_aggregated['Дата']).dt.day

daily_aggregated = daily_aggregated.sort_values('Дата')
```

### Подготовка данных для моделирования временных рядов

```python
df = df_new.copy()
```

```python
df['target'] = df['num_orders'].shift(-1)
```

```python
df = df.iloc[:-1]
```

```python
# Создаю лаги: за 1, 2, 3, 6, 12, 24 часа, а также за неделю
for lag in [1, 2, 3, 6, 12, 24, 168]:
    df[f'lag_{lag}'] = df['num_orders'].shift(lag)
```

```python
# Создаю скользящие статистики
for window in [3, 6, 12, 24]:
    df[f'rolling_mean_{window}'] = df['num_orders'].rolling(window=window, min_periods=1).mean().shift(1)
    df[f'rolling_std_{window}'] = df['num_orders'].rolling(window=window, min_periods=1).std().shift(1)
```

```python
# Добавляю признак "час в сутках" в виде синуса/косинуса для цикличности
df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
df['day_of_week_sin'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
df['day_of_week_cos'] = np.cos(2 * np.pi * df['day_of_week'] / 7)
```

```python
# Удаляю строки с пропусками из-за лагов
df = df.dropna().reset_index(drop=True)
```

```python
# Определяю признаки и целевую переменную
features = [
    'hour', 'month', 'day_of_week', 'day_of_month', 'is_weekend', 'is_holiday',
    'hour_sin', 'hour_cos', 'day_of_week_sin', 'day_of_week_cos',
    'lag_1', 'lag_2', 'lag_3', 'lag_6', 'lag_12', 'lag_24', 'lag_168',
    'rolling_mean_3', 'rolling_std_3', 'rolling_mean_6', 'rolling_std_6',
    'rolling_mean_12', 'rolling_std_12', 'rolling_mean_24', 'rolling_std_24'
]

X = df[features]
y = df['target']
```

### Функция для оценки модели TimeSeriesSplit

```python
def evaluate_model_tscv(model, model_name, X, y, param_grid=None):

    print(f"\n{'='*60}")
    print(f"МОДЕЛЬ: {model_name}")
    print(f"{'='*60}")

    results = {'model_name': model_name}

    # КРОСС-ВАЛИДАЦИЯ С TIMESERIESSPLIT
    print("\n. КРОСС-ВАЛИДАЦИЯ TimeSeriesSplit:")

    start = time.time()

    if param_grid:
        print(f"  Поиск лучших гиперпараметров")
        grid_search = GridSearchCV(
            estimator=model,
            param_grid=param_grid,
            cv=tscv,
            scoring='neg_mean_squared_error',
            n_jobs=-1,
            verbose=0
        )

        grid_search.fit(X, y)

        best_model = grid_search.best_estimator_
        best_params = grid_search.best_params_
        cv_rmse = np.sqrt(-grid_search.best_score_)

        print(f"  Лучшие параметры: {best_params}")
        print(f"  Лучший CV RMSE: {cv_rmse:.2f}")

        results['best_params'] = best_params
        results['cv_rmse'] = cv_rmse
        results['best_model'] = best_model

    else:
        cv_scores = cross_val_score(
            model,
            X,
            y,
            cv=tscv,
            scoring='neg_mean_squared_error',
            n_jobs=-1
        )

        cv_time = time.time() - start
        cv_rmse_scores = np.sqrt(-cv_scores)
        cv_rmse = cv_rmse_scores.mean()
        cv_std = cv_rmse_scores.std()

        print(f"  CV время: {cv_time:.2f} сек")
        print(f"  CV RMSE: {cv_rmse:.2f} ± {cv_std:.2f}")
        print(f"  CV scores по фолдам: {cv_rmse_scores.round(2)}")

        results['cv_rmse'] = cv_rmse
        results['cv_std'] = cv_std
        results['cv_time'] = cv_time
        results['cv_scores'] = cv_rmse_scores.tolist()

    print("\n. ОБУЧЕНИЕ МОДЕЛИ:")

    # Разделяем на train/test (последние 10% - тест)
    test_size = int(len(X) * 0.1)
    X_train = X.iloc[:-test_size]
    y_train = y.iloc[:-test_size]
    X_test = X.iloc[-test_size:]
    y_test = y.iloc[-test_size:]

    print(f"  Размер train: {len(X_train)}")
    print(f"  Размер test: {len(X_test)}")

    # Масштабирование признаков
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    # Обучение модели
    start_train = time.time()
    if param_grid and 'best_model' in results:
        model_to_train = results['best_model']
    else:
        model_to_train = model

    model_to_train.fit(X_train_scaled, y_train)
    train_time = time.time() - start_train

    print(f"  Время обучения: {train_time:.2f} сек")

    # ПРЕДСКАЗАНИЯ И ОЦЕНКА
    print("\n3. ОЦЕНКА МОДЕЛИ:")

    # На тренировочных данных
    train_pred = model_to_train.predict(X_train_scaled)
    train_rmse = np.sqrt(mean_squared_error(y_train, train_pred))
    print(f"  Train RMSE: {train_rmse:.2f}")

    # На тестовых данных
    test_pred = model_to_train.predict(X_test_scaled)
    test_rmse = np.sqrt(mean_squared_error(y_test, test_pred))
    print(f"  Test RMSE: {test_rmse:.2f}")

    # Сохраняем предсказания для визуализации
    results['test_predictions'] = test_pred

    # Проверка целевой метрики
    target_rmse = 48
    meets_target = test_rmse < target_rmse
    print(f"  Соответствует RMSE < {target_rmse}: {'ДА ✅' if meets_target else 'НЕТ ❌'}")

    results.update({
        'train_rmse': train_rmse,
        'test_rmse': test_rmse,
        'train_time': train_time,
        'meets_target': meets_target,
        'model': model_to_train
    })

    overfit_score = train_rmse / test_rmse
    print(f"  Отношение Train/Test RMSE: {overfit_score:.3f}")
    if overfit_score < 0.8:
        print("Возможное недообучение (train ошибка значительно больше test)")
    elif overfit_score > 1.2:
        print(" Возможное переобучение (train ошибка значительно меньше test)")
    else:
        print("Модель хорошо обобщается")

    return results
```

### Обучение и оценка разных моделей с TimeSeriesSplit

```python
all_results_tscv = []

# LINEAR REGRESSION
print("\n" + "="*60)
lr_results = evaluate_model_tscv(
    model=LinearRegression(),
    model_name='LinearRegression',
    X=X,
    y=y
)
all_results_tscv.append(lr_results)

# RIDGE REGRESSION
print("\n" + "="*60)
ridge_param_grid = {
    'alpha': [0.001, 0.01, 0.1, 1, 10, 100]
}

ridge_results = evaluate_model_tscv(
    model=Ridge(random_state=42),
    model_name='RidgeRegression',
    X=X,
    y=y,
    param_grid=ridge_param_grid
)
all_results_tscv.append(ridge_results)

# LASSO REGRESSION
print("\n" + "="*60)
lasso_param_grid = {
    'alpha': [0.0001, 0.001, 0.01, 0.1, 1]
}

lasso_results = evaluate_model_tscv(
    model=Lasso(random_state=42, max_iter=10000),
    model_name='LassoRegression',
    X=X,
    y=y,
    param_grid=lasso_param_grid
)
all_results_tscv.append(lasso_results)

# RANDOM FOREST
print("\n" + "="*60)
rf_param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [10, 20],
    'min_samples_split': [2, 5]
}

rf_results = evaluate_model_tscv(
    model=RandomForestRegressor(random_state=42, n_jobs=-1),
    model_name='RandomForest',
    X=X,
    y=y,
    param_grid=rf_param_grid
)
all_results_tscv.append(rf_results)

# GRADIENT BOOSTING
gb_param_grid = {
    'n_estimators': [50, 100],
    'learning_rate': [0.01, 0.1],
    'max_depth': [3, 5]
}

gb_results = evaluate_model_tscv(
    model=GradientBoostingRegressor(random_state=42),
    model_name='GradientBoosting',
    X=X,
    y=y,
    param_grid=gb_param_grid
)
all_results_tscv.append(gb_results)
```
